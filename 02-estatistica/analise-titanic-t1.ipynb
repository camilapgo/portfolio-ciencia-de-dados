## <font color = Red> **2º INFORMAÇÃO RELEVANTE**
**SOBREVIVENCIA POR PERFIS DE CLASSE SOCIAL**
****

## CÓDIGO DOIS

```python
classes = ['First', 'Second', 'Third']
classes_ptbr = {'First': 'Primeira', 'Second': 'Segunda', 'Third': 'Terceira'}
sobreviventes_classe = []
perc_sobreviventes_classe = []
total_classe = []
# calculo da % de cada sobrevivencia por classe
for classe in classes:
    passageiros_classe = df[df['class'] == classe]
    total_classe.append(len(passageiros_classe))
    vivos = len(passageiros_classe[passageiros_classe['survived'] == 1])
    sobreviventes_classe.append(vivos)
    perc = (vivos / len(passageiros_classe)) * 100
    perc_sobreviventes_classe.append(perc)
    print(f"{classes_ptbr[classe]} classe: {len(passageiros_classe)} passageiros, {vivos} sobreviventes ({perc:.2f}%)")

# relação de sobrevivencia dos homens e mulheres dentro de cada classe:
for classe in classes:
    df_classe = df[df['class'] == classe]
    homens_classe = df_classe[df_classe['sex'] == 'male']
    mulheres_classe = df_classe[df_classe['sex'] == 'female']
    homens_vivos_classe = len(homens_classe[homens_classe['survived'] == 1])
    mulheres_vivas_classe = len(mulheres_classe[mulheres_classe['survived'] == 1])
    perc_homens_classe = (homens_vivos_classe / len(homens_classe)) * 100
    perc_mulheres_classe = (mulheres_vivas_classe / len(mulheres_classe)) * 100
    print(f"\n{classes_ptbr[classe]} classe:")
    print(f"  Homens: {homens_vivos_classe} sobreviventes ({perc_homens_classe:.2f}% dos homens da classe)")
    print(f"  Mulheres: {mulheres_vivas_classe} sobreviventes ({perc_mulheres_classe:.2f}% das mulheres da classe)")
```

A taxa de sobrevivência na Primeira classe (62.96%) foi mais que o dobro da Terceira classe (24.24%). Ou seja, 136 pessoas da Primeira classe sobreviveram, contra 119 da Terceira, mesmo que a Terceira tenha tido muito mais passageiros.

## GRÁFICO DOIS

```python
# GRÁFICO DA TAXA DE SOBREVIVENCIA POR CLASSE
plt.subplot(1,2,1)
classes_nomes = ['Primeira', 'Segunda', 'Terceira']
cores2 = ['#ff9999', '#66b3ff', '#99ff99']
sns.barplot(x=classes_nomes, y=perc_sobreviventes_classe, hue=classes_nomes, edgecolor='black', palette=cores2)
plt.title('Sobrevivência por Classe Social', fontsize=12, fontweight='bold', pad=15)
plt.ylabel('Taxa de Sobrevivência (%)', fontweight='bold', fontsize=10)
plt.xlabel('Classe', fontweight='bold', fontsize=10)
plt.ylim(0,100)
plt.grid(axis='y', linestyle='--')
```

```python
# GRAFICO DE BARRAS DO NUMERO DE SOBREVIVENTES VS MORTOS POR CLASSE
sobreviventes_classe = []
mortos_classe = []
for classe in classes:
    sobreviventes_classe.append(len(df[(df['class']==classe) & (df['survived']==1)]))
    mortos_classe.append(len(df[(df['class']==classe) & (df['survived']==0)]))
plt.figure(figsize=(8,6))
x = [0, 1, 2]  # posições para as três classes
largura = 0.3
plt.bar([p - largura/2 for p in x], sobreviventes_classe, largura, label='Sobreviventes', edgecolor='black', color='#99ff99')
plt.bar([p + largura/2 for p in x], mortos_classe, largura, label='Mortos', edgecolor='black', color='#ff9999')
plt.xticks(x, ['Primeira', 'Segunda', 'Terceira'])
plt.xlabel('Classe', fontweight='bold')
plt.ylabel('Número de passageiros', fontweight='bold')
plt.title('Comparação de Sobreviventes e Mortos por Classe', fontsize=14, fontweight='bold', pad=15)
plt.legend()
plt.grid(axis='y', linestyle='--')
```

```python
sobreviventes = df[df['survived']==1]
mortos = df[df['survived']==0]
sobreviventes_juntos = sobreviventes[sobreviventes['alone']==False]
mortos_juntos = mortos[mortos['alone']==False]
sobreviventes_sozinhos = sobreviventes[sobreviventes['alone']==True]
mortos_sozinhos = mortos[mortos['alone']==True]
total_sobreviventes_juntos = len(sobreviventes_juntos)
total_sobreviventes_sozinhos = len(sobreviventes_sozinhos)
total_mortos_juntos = len(mortos_juntos)
total_mortos_sozinhos = len(mortos_sozinhos)

# PERCENTUAL DE CADA TIPO
perc_sobreviventes_juntos = round(total_sobreviventes_juntos*100/len(sobreviventes),2)
perc_mortos_juntos = round(total_mortos_juntos*100/len(mortos),2)
perc_sobreviventes_sozinhos = round(total_sobreviventes_sozinhos*100/len(sobreviventes),2)
perc_mortos_sozinhos = round(total_mortos_sozinhos*100/len(mortos),2)

print(f'O número de sobreviventes eram de: {len(sobreviventes)}')
print(f'O número de mortos eram de: {len(mortos)}')
print(f'O número de sobreviventes que estavam juntos com outras pessoas eram {total_sobreviventes_juntos} e sozinhos eram {total_sobreviventes_sozinhos}')
print(f'O número de mortos que estavam juntos com outras pessoas eram {total_mortos_juntos} e sozinhos eram {total_mortos_sozinhos}')
print(f'\nO percentual de cada tópico é de:\nSobreviventes juntos: {perc_sobreviventes_juntos}%')
print(f'Sobreviventes sozinhos: {perc_sobreviventes_sozinhos}%')
print(f'Mortos juntos: {perc_mortos_juntos}%')
print(f'Mortos sozinhos: {perc_mortos_sozinhos}%')
```

## <font color = Red> **4º INFORMAÇÃO RELEVANTE**
**ANÁLISE DO PREÇO DA PASSAGEM (FARE) E SOBREVIVÊNCIA**
****

## CÓDIGO QUATRO

```python
# calculando o fare de acordo com a taxa de sobrevivencia
fare_sobreviventes = df[df['survived'] == 1]['fare']
fare_mortos = df[df['survived'] == 0]['fare']
media_fare_vivos = fare_sobreviventes.mean()
media_fare_mortos = fare_mortos.mean()
mediana_fare_vivos = fare_sobreviventes.median()
mediana_fare_mortos = fare_mortos.median()
print("Comparação do Valor da Passagem:")
print(f"Sobreviventes: média = £{media_fare_vivos:.2f}, mediana = £{mediana_fare_vivos:.2f}")
print(f"Mortos: média = £{media_fare_mortos:.2f}, mediana = £{mediana_fare_mortos:.2f}")
print(f"Diferença média: £{media_fare_vivos - media_fare_mortos:.2f} a mais para os sobreviventes")

# Fare + Classe
classes_ptbr = {'First': 'Primeira', 'Second': 'Segunda', 'Third': 'Terceira'}
print("\nAnálise: Preço da passagem dentro da mesma classe")
for classe in classes_ptbr:
    df_classe = df[df['class'] == classe]
    fare_mediana = df_classe['fare'].median()
    acima_mediana = df_classe[df_classe['fare'] > fare_mediana]
    abaixo_mediana = df_classe[df_classe['fare'] <= fare_mediana]
    taxa_acima = (len(acima_mediana[acima_mediana['survived'] == 1]) / len(acima_mediana)) * 100
    taxa_abaixo = (len(abaixo_mediana[abaixo_mediana['survived'] == 1]) / len(abaixo_mediana)) * 100
    print(f"{classes_ptbr[classe]} classe:")
    print(f"  Quem pagou acima da mediana ({fare_mediana:.2f}): {taxa_acima:.1f}% de sobrevivência")
    print(f"  Quem pagou abaixo da mediana: {taxa_abaixo:.2f}% de sobrevivência")
    print(f"  Diferença: {taxa_acima - taxa_abaixo:.2f}%")
    print(f"\n")
```

```python
df_dead = df[(df["survived"] == 0)]
q1_dead = df_dead.fare.quantile(0.25)
q3_dead = df_dead.fare.quantile(0.75)
AIQ_dead = q3_dead - q1_dead

LI_dead = q1_dead - 1.5 * AIQ_dead
LS_dead = q3_dead + 1.5 * AIQ_dead

dead_no_outliers = df_dead[(df_dead.fare >= LI_dead) & (df_dead.fare <= LS_dead)]

dead_no_outliers.fare.max() #-> valor máximo de fare dos mortos, sem outliers
```

## <font color = Red> **5º INFORMAÇÃO RELEVANTE**
**ANÁLISE DE TAMANHO DE FAMÍLIA COM TAXA DE SOBREVIVENCIA**
****

## CÓDIGO CINCO

```python
df['family_size'] = df['sibsp'] + df['parch'] + 1

# Classificando em grupos de tamanho
sozinho_familia = df[df['family_size'] == 1]
pequena = df[(df['family_size'] >= 2) & (df['family_size'] <= 3)]
media = df[(df['family_size'] >= 4) & (df['family_size'] <= 6)]
grande = df[df['family_size'] >= 7]

# Calculando taxas de sobrevivência para cada grupo
taxa_sozinho = len(sozinho_familia[sozinho_familia['survived'] == 1]) * 100 / len(sozinho_familia)
taxa_pequena = len(pequena[pequena['survived'] == 1]) * 100 / len(pequena)
taxa_media = len(media[media['survived'] == 1]) * 100 / len(media)
taxa_grande = len(grande[grande['survived'] == 1]) * 100 / len(grande)

print('ANÁLISE POR TAMANHO DA FAMÍLIA:')
print(f'Tamanho "Sozinho" (1 pessoa): {taxa_sozinho:.2f}% de sobrevivência')
print(f'Tamanho "Pequena" (2-3 pessoas): {taxa_pequena:.2f}% de sobrevivência')
print(f'Tamanho "Média" (4-6 pessoas): {taxa_media:.2f}% de sobrevivência')
print(f'Tamanho "Grande" (7+ pessoas): {taxa_grande:.2f}% de sobrevivência')
```

## GRÁFICO CINCO

```python
categorias = ['Sozinho (1)', 'Pequena (2-3)', 'Média (4-6)', 'Grande (7+)']
taxas = [taxa_sozinho, taxa_pequena, taxa_media, taxa_grande]
plt.figure(figsize=(8, 5))
cores = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
barras = plt.bar(categorias, taxas, color=cores, edgecolor='black', linewidth=1.2)
plt.title('Taxa de Sobrevivência por Tamanho da Família', fontsize=14, fontweight='bold', pad=15)
plt.ylabel('Taxa de Sobrevivência (%)', fontsize=11, fontweight='bold')
plt.xlabel('Tamanho da Família', fontsize=11, fontweight='bold')
plt.ylim(0, max(taxas) + 10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
