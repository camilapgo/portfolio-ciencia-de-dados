## FUNÇÕES UTILIZADAS PESQUISADAS
`.strip()` ->  remove espaços em branco ou caracteres específicos do início e do final de uma string.  
`random.randint` ->  gera números inteiros aleatórios dentro de um intervalo definido.  
`random.shuffle` ->  altera aleatoriamente a ordem dos elementos da lista original e não retorna uma nova lista.  
`isalpha()` ->  verifica se uma string contém apenas caracteres alfabéticos.  
`.upper()` ->  converte texto em letras maiúsculas.  
`abs()` ->  retorna o valor absoluto de um número.  
`str()` ->  converte um objeto em uma string.  
`datetime.now().strftime()` ->  obtém a data e hora atuais e formata como uma string.  
`encoding="utf-8"` -> evita problemas de portabilidade, garantindo que caracteres especiais sejam tratados corretamente

```python
# IMPRIMIR INTERFACE JOGO
print("==============================================================")
print("  jjjjjjj   ooooo    ggggg    ooooo       dddd    eeeee       ")
print("      j    o     o  g        o     o      d   d   e           ")
print("      j    o     o  g  gggg  o     o      d    d  eeeee       ")
print("  j   j    o     o  g     g  o     o      d   d   e           ")
print("   jjj      ooooo    ggggg    ooooo       dddd    eeeee       ")
print("\n")
print("   aaaa   ddddd    i  v     v  i  nnn n  h  h   aaaa   rrrr  ")
print("  a    a  d    d   i  vv   vv  i  n n n  h  h  a    a  r  r  ")
print("  aaaaaa  d     d  i   vv vv   i  n n n  hhhh  aaaaaa  r r   ")
print("  a    a  d    d   i    vvv    i  n n n  h  h  a    a  r  r  ")
print("  a    a  ddddd    i     v     i  n nnn  h  h  a    a  r   r ")
print("==============================================================")

# NOME DO JOGADOR
nome_do_jogador = ""
while nome_do_jogador == "":
    nome_do_jogador = input("Digite seu nome para começar: ").strip()
    if nome_do_jogador == "":
        print("\n")
        print("O nome não pode estar em branco.")

# MENU PRINCIPAL
jogo_esta_ativo = True
while jogo_esta_ativo:
    print("\n")
    print("==============================================================")
    print("        PALAVRA SECRETA - MITOLOGIA GREGA")
    print("==============================================================")
    print("  Jogador:", nome_do_jogador)
    print("--------------------------------------------------------------")
    print("  [1]  Jogar nova partida")
    print("  [2]  Ver histórico de partidas")
    print("  [3]  Como jogar")
    print("  [4]  Sair")
    print("--------------------------------------------------------------")
    opcao_do_menu = input("Escolha: ").strip()

    # OPCAO 1 - JOGAR NOVA PARTIDA
    if opcao_do_menu == "1":

        # Escolher o nivel de dificuldade
        nivel_foi_escolhido = False
        nivel_escolhido     = ""
        while not nivel_foi_escolhido:
            print("\n")
            print("--------------------------------------------------------------")
            print("  Escolha o nível:")
            print("  [1] FÁCIL    - 10 tentativas | 2 letras reveladas | +10 pts")
            print("  [2] MÉDIO    -  7 tentativas | 1 letra  revelada  | +15 pts")
            print("  [3] DIFÍCIL  -  5 tentativas | sem ajuda          | +20 pts")
            print("  [0] Voltar ao menu principal")
            print("--------------------------------------------------------------")
            opcao_de_nivel = input("  Nivel: ").strip()

            if opcao_de_nivel == "0":
                nivel_foi_escolhido = True

            elif opcao_de_nivel == "1":
                nivel_escolhido            = "FACIL"
                tentativas_maximas         = 10
                quantidade_letras_visiveis = 2
                pontos_por_acerto          = 10
                pontos_por_erro            = -5
                nivel_foi_escolhido        = True

            elif opcao_de_nivel == "2":
                nivel_escolhido            = "MEDIO"
                tentativas_maximas         = 7
                quantidade_letras_visiveis = 1
                pontos_por_acerto          = 15
                pontos_por_erro            = -7
                nivel_foi_escolhido        = True

            elif opcao_de_nivel == "3":
                nivel_escolhido            = "DIFICIL"
                tentativas_maximas         = 5
                quantidade_letras_visiveis = 0
                pontos_por_acerto          = 20
                pontos_por_erro            = -10
                nivel_foi_escolhido        = True

            else:
                print("\n")
                print("Opção inválida. Escolha 1, 2, 3 ou 0.")

        # Só continua se escolheu um nível
        if nivel_escolhido != "":

            # Filtrar palavras pelo nível escolhido
            palavras_do_nivel = []

            for item_do_banco in BANCO:
                if item_do_banco[2] == nivel_escolhido:
                    palavras_do_nivel.append(item_do_banco)

            if len(palavras_do_nivel) == 0:
                palavras_do_nivel = BANCO

            # Sortear uma palavra aleatoria
            indice_sorteado  = random.randint(0, len(palavras_do_nivel) - 1)
            item_sorteado    = palavras_do_nivel[indice_sorteado]
            palavra_original     = item_sorteado[0]
            dica_da_palavra      = item_sorteado[1]
            nivel_da_palavra     = item_sorteado[2]
            categoria_da_palavra = item_sorteado[3]

            # Remover acentos
            palavra_sem_acento = ""
            for caractere_original in palavra_original.upper():
                if caractere_original == "Á":
                    letra_sem_acento = "A"
                elif caractere_original == "À":
                    letra_sem_acento = "A"
                elif caractere_original == "Â":
                    letra_sem_acento = "A"
                elif caractere_original == "Ã":
                    letra_sem_acento = "A"
                elif caractere_original == "É":
                    letra_sem_acento = "E"
                elif caractere_original == "È":
                    letra_sem_acento = "E"
                elif caractere_original == "Ê":
                    letra_sem_acento = "E"
                elif caractere_original == "Í":
                    letra_sem_acento = "I"
                elif caractere_original == "Ì":
                    letra_sem_acento = "I"
                elif caractere_original == "Î":
                    letra_sem_acento = "I"
                elif caractere_original == "Ó":
                    letra_sem_acento = "O"
                elif caractere_original == "Ò":
                    letra_sem_acento = "O"
                elif caractere_original == "Ô":
                    letra_sem_acento = "O"
                elif caractere_original == "Õ":
                    letra_sem_acento = "O"
                elif caractere_original == "Ú":
                    letra_sem_acento = "U"
                elif caractere_original == "Ù":
                    letra_sem_acento = "U"
                elif caractere_original == "Û":
                    letra_sem_acento = "U"
                elif caractere_original == "Ç":
                    letra_sem_acento = "C"
                else:
                    letra_sem_acento = caractere_original
                palavra_sem_acento = palavra_sem_acento + letra_sem_acento

            # Escolher quais posicoes aparecerao reveladas desde o inicio da partida
            indices_validos = []
            for indice_da_posicao in range(len(palavra_sem_acento)):
                letra_nessa_posicao = palavra_sem_acento[indice_da_posicao]

                if letra_nessa_posicao != " " and letra_nessa_posicao != "-":
                    indices_validos.append(indice_da_posicao)

            random.shuffle(indices_validos)

            indices_das_posicoes_visiveis = []
            for numero_do_visivel in range(quantidade_letras_visiveis):
                if numero_do_visivel < len(indices_validos):
                    indices_das_posicoes_visiveis.append(indices_validos[numero_do_visivel])

            # Contadores e Listas
            tentativas_restantes = tentativas_maximas
            pontuacao_atual      = 0
            numero_de_acertos    = 0
            numero_de_erros      = 0
            letras_acertadas     = []
            letras_erradas       = []

            # As letras das posicoes visiveis ja entram nas acertadas
            for indice_visivel in indices_das_posicoes_visiveis:
                letra_visivel = palavra_sem_acento[indice_visivel]
                letra_ja_esta_na_lista = False

                for letra_acertada in letras_acertadas:
                    if letra_acertada == letra_visivel:
                        letra_ja_esta_na_lista = True
                        break

                if not letra_ja_esta_na_lista:
                    letras_acertadas.append(letra_visivel)

            # JOGO PRINCIPAL
            while tentativas_restantes > 0:

                exibicao_da_palavra = ""

                for indice_da_posicao in range(len(palavra_sem_acento)):
                    letra_na_posicao = palavra_sem_acento[indice_da_posicao]

                    if letra_na_posicao == " " or letra_na_posicao == "-":
                        exibicao_da_palavra = exibicao_da_palavra + "  "

                    else:
                        letra_foi_acertada = False

                        for letra_acertada in letras_acertadas:

                            if letra_acertada == letra_na_posicao:
                                letra_foi_acertada = True
                                break

                        posicao_e_visivel = False

                        for indice_visivel in indices_das_posicoes_visiveis:

                            if indice_visivel == indice_da_posicao:
                                posicao_e_visivel = True
                                break

                        if letra_foi_acertada or posicao_e_visivel:
                            exibicao_da_palavra = exibicao_da_palavra + letra_na_posicao + " "

                        else:
                            exibicao_da_palavra = exibicao_da_palavra + "_ "

                # Painel do jogo
                print("\n")
                print("==============================================================")
                print("        PALAVRA SECRETA - MITOLOGIA GREGA")
                print("==============================================================")
                print("  Jogador  :", nome_do_jogador)
                print("  Categoria:", categoria_da_palavra, " | Nivel:", nivel_da_palavra)
                print("--------------------------------------------------------------")
                print("  Dica:", dica_da_palavra)
                print("\n")

                # Exibir a palavra com letras bem espaçadas
                palavra_para_mostrar = ""

                for letra_exibida in exibicao_da_palavra.split():
                    palavra_para_mostrar = palavra_para_mostrar + letra_exibida + "  "

                print("\t", palavra_para_mostrar.strip(), "\n")

                # Lista de letras ja erradas
                texto_das_letras_erradas = ""

                for letra_errada in letras_erradas:
                    texto_das_letras_erradas = texto_das_letras_erradas + letra_errada + " "

                if texto_das_letras_erradas == "":
                    texto_das_letras_erradas = "---"

                print("  Letras erradas :", texto_das_letras_erradas)
                print("--------------------------------------------------------------")
                print("  Pontuação  :", pontuacao_atual)
                texto_acertos_e_erros = "  Acertos : " + str(numero_de_acertos) + "  |  Erros : " + str(numero_de_erros)
                print(texto_acertos_e_erros)

                # Barra de tentativas
                barra_de_tentativas = ""

                for numero_barra in range(tentativas_restantes):
                    barra_de_tentativas = barra_de_tentativas + "#"

                for numero_barra in range(tentativas_maximas - tentativas_restantes):
                    barra_de_tentativas = barra_de_tentativas + "."

                texto_contagem = str(tentativas_restantes) + "/" + str(tentativas_maximas)
                linha_de_tentativas = "  Tentativas : [" + barra_de_tentativas + "] " + texto_contagem

                print(linha_de_tentativas)
                print("--------------------------------------------------------------")

                # Verificar se já revelou a palavra toda
                palavra_esta_completa = True

                for indice_da_posicao in range(len(palavra_sem_acento)):
                    letra_na_posicao = palavra_sem_acento[indice_da_posicao]

                    if letra_na_posicao != " " and letra_na_posicao != "-":

                        letra_foi_acertada = False
                        for letra_acertada in letras_acertadas:
                            if letra_acertada == letra_na_posicao:
                                letra_foi_acertada = True
                                break

                        posicao_e_visivel = False
                        for indice_visivel in indices_das_posicoes_visiveis:
                            if indice_visivel == indice_da_posicao:
                                posicao_e_visivel = True
                                break

                        if not letra_foi_acertada and not posicao_e_visivel:
                            palavra_esta_completa = False
                            break

                if palavra_esta_completa:
                    break

                # Ler e validar a letra digitada
                letra_digitada_e_valida = False

                while not letra_digitada_e_valida:

                    entrada_digitada = input("  Digite uma letra: ").strip()
                    if len(entrada_digitada) == 0:
                        print("\n")
                        print("Digite ao menos uma letra.")

                    elif len(entrada_digitada) > 1:
                        print("\n")
                        print("Digite apenas UMA letra por vez.")

                    elif not entrada_digitada.isalpha():
                        print("\n")
                        print("Digite apenas letras do alfabeto.")

                    else:
                        letra_digitada = entrada_digitada.upper()

                        # Remover acento da letra digitada
                        if letra_digitada == "Á":
                            letra_digitada = "A"
                        elif letra_digitada == "À":
                            letra_digitada = "A"
                        elif letra_digitada == "Â":
                            letra_digitada = "A"
                        elif letra_digitada == "Ã":
                            letra_digitada = "A"
                        elif letra_digitada == "É":
                            letra_digitada = "E"
                        elif letra_digitada == "È":
                            letra_digitada = "E"
                        elif letra_digitada == "Ê":
                            letra_digitada = "E"
                        elif letra_digitada == "Í":
                            letra_digitada = "I"
                        elif letra_digitada == "Ì":
                            letra_digitada = "I"
                        elif letra_digitada == "Î":
                            letra_digitada = "I"
                        elif letra_digitada == "Ó":
                            letra_digitada = "O"
                        elif letra_digitada == "Ò":
                            letra_digitada = "O"
                        elif letra_digitada == "Ô":
                            letra_digitada = "O"
                        elif letra_digitada == "Õ":
                            letra_digitada = "O"
                        elif letra_digitada == "Ú":
                            letra_digitada = "U"
                        elif letra_digitada == "Ù":
                            letra_digitada = "U"
                        elif letra_digitada == "Û":
                            letra_digitada = "U"
                        elif letra_digitada == "Ç":
                            letra_digitada = "C"

                        # Verificar se essa letra ja foi tentada
                        letra_ja_foi_tentada = False

                        for letra_errada in letras_erradas:
                            if letra_errada == letra_digitada:
                                letra_ja_foi_tentada = True
                                break

                        for letra_acertada in letras_acertadas:
                            if letra_acertada == letra_digitada:
                                letra_ja_foi_tentada = True
                                break

                        if letra_ja_foi_tentada:
                            print("\n")
                            print("Você já tentou a letra", letra_digitada, ". Escolha outra.")
                        else:
                            letra_digitada_e_valida = True

                # Verificar erro/acerto/repetição da letra
                numero_de_ocorrencias = 0

                for caractere_da_palavra in palavra_sem_acento:
                    if caractere_da_palavra == letra_digitada:
                        numero_de_ocorrencias = numero_de_ocorrencias + 1

                if numero_de_ocorrencias > 0:
                    letras_acertadas.append(letra_digitada)
                    ganho_de_pontos   = pontos_por_acerto * numero_de_ocorrencias
                    pontuacao_atual   = pontuacao_atual + ganho_de_pontos
                    numero_de_acertos = numero_de_acertos + 1
                    print("\n")
                    print("CERTO! A letra", letra_digitada, "aparece", numero_de_ocorrencias, "vez(es).")
                    print("Você ganhou", ganho_de_pontos, "pontos!")

                else:
                    letras_erradas.append(letra_digitada)
                    pontuacao_atual      = pontuacao_atual + pontos_por_erro
                    tentativas_restantes = tentativas_restantes - 1
                    numero_de_erros      = numero_de_erros + 1
                    print("\n")
                    print("ERROU! A letra", letra_digitada, "nao esta na palavra.")
                    print("Voce perdeu", abs(pontos_por_erro), "pontos.")

                input("[ ENTER para continuar ]")

            # RESULTADO DA PARTIDA
            jogador_venceu = True

            for indice_da_posicao in range(len(palavra_sem_acento)):
                letra_na_posicao = palavra_sem_acento[indice_da_posicao]

                if letra_na_posicao != " " and letra_na_posicao != "-":

                    letra_foi_acertada = False
                    for letra_acertada in letras_acertadas:
                        if letra_acertada == letra_na_posicao:
                            letra_foi_acertada = True
                            break

                    posicao_e_visivel = False
                    for indice_visivel in indices_das_posicoes_visiveis:
                        if indice_visivel == indice_da_posicao:
                            posicao_e_visivel = True
                            break

                    if not letra_foi_acertada and not posicao_e_visivel:
                        jogador_venceu = False
                        break

            print("\n==============================================================")

            if jogador_venceu:
                pontuacao_atual = pontuacao_atual + 50
                print("\t\t\tPARABENS! VOCÊ VENCEU!")
                print("\t\t\tBonus de vitória: +50 pontos!")

            else:
                print("\t\t\tSUAS TENTATIVAS ACABARAM...")
                print("\t\t\tA palavra era:", palavra_original)

            numero_de_tentativas_usadas = tentativas_maximas - tentativas_restantes
            print("--------------------------------------------------------------")
            print("  Palavra secreta   :", palavra_original)
            print("  Pontuação final   :", pontuacao_atual)
            print("  Acertos           :", numero_de_acertos)
            print("  Erros             :", numero_de_erros)

            texto_tentativas_usadas = str(numero_de_tentativas_usadas) + "/" + str(tentativas_maximas)
            print("  Tentativas usadas :", texto_tentativas_usadas)
            print("==============================================================")

            # Salvar o resultado no arquivo
            data_e_hora = datetime.now().strftime("%d/%m/%Y %H:%M")

            if jogador_venceu:
                texto_do_resultado = "VENCEU"

            else:
                texto_do_resultado = "PERDEU"

            linha_para_salvar = (data_e_hora + " | " +
                                 "Jogador: "     + nome_do_jogador + " | " +
                                 "Pontos: "      + str(pontuacao_atual) + " | " +
                                 "Tentativas: "  + str(numero_de_tentativas_usadas) + "/" + str(tentativas_maximas) + " | " +
                                 "Nivel: "       + nivel_da_palavra + " | " + texto_do_resultado + " | " +
                                 "Palavra: "     + palavra_original + "\n")

            arquivo_de_resultados = open("resultados.txt", "a", encoding="utf-8")
            arquivo_de_resultados.write(linha_para_salvar)
            arquivo_de_resultados.close()

            input("[ ENTER para continuar ]")

    # OPÇÃO 2 - VER HISTORICO DE PARTIDAS
    elif opcao_do_menu == "2":
        print("\n")
        print("==============================================================")
        print("            HISTÓRICO DE PARTIDAS")
        print("==============================================================")

        try:
            arquivo_historico = open("resultados.txt", "r", encoding="utf-8")
            numero_de_linhas_lidas = 0
            linha_lida = arquivo_historico.readline()

            while linha_lida != "":
                print(" ", linha_lida.strip())
                numero_de_linhas_lidas = numero_de_linhas_lidas + 1
                linha_lida = arquivo_historico.readline()

            arquivo_historico.close()

            if numero_de_linhas_lidas == 0:
                print("  Nenhuma partida registrada ainda.")

        except:
            print("  Nenhuma partida registrada ainda.")

        input("  [ ENTER para continuar ]")

    # OPÇÃO 3 - COMO JOGAR
    elif opcao_do_menu == "3":
        print("\n")
        print("==============================================================")
        print("                    COMO JOGAR")
        print("==============================================================")
        print("\n")
        print("  OBJETIVO")
        print("  Descubra a palavra da Mitologia Grega antes")
        print("  de esgotar todas as tentativas!")
        print("\n")
        print("  COMO FUNCIONA")
        print("  1. O computador sorteia uma palavra e mostra uma DICA.")
        print("  2. Os _ representam as letras ainda não descobertas.")
        print("  3. Digite UMA letra por vez e pressione ENTER.")
        print("  4. Letra CERTA  -> aparece na palavra e você ganha pontos.")
        print("  5. Letra ERRADA -> você perde pontos e uma tentativa.")
        print("  6. Algumas letras já aparecem reveladas (depende do nível).")
        print("\n")
        print("  NÍVEIS")
        print("  FÁCIL   : 10 tentativas | 2 letras reveladas | +10 pts")
        print("  MÉDIO   :  7 tentativas | 1 letra  revelada  | +15 pts")
        print("  DIFÍCIL :  5 tentativas | sem ajuda           | +20 pts")
        print("\n")
        print("  BÔNUS: +50 pontos ao vencer a partida!")
        print("\n")
        input("  [ ENTER para voltar ]")

    # OPÇÃO 4 - SAIR
    elif opcao_do_menu == "4":
        print("\n")
        print("==============================================================")
        print("Obrigado por jogar,", nome_do_jogador)
        print("Que os deuses do Olimpo estejam com você!")
        print("==============================================================")
        jogo_esta_ativo = False

    else:
        print("\n")
        print("Opção inválida. Escolha entre 1, 2, 3 ou 4.")
```
